Предлагаю отобразить баллы, набранные командами, которые заняли 18, 19 и 20 места.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('eplleaguetables.csv')
data = data[np.logical_or(data["Position"] == 18, np.logical_or(data["Position"] == 19, data["Position"] == 20))]
data.sort_values("Points")
data.boxplot(column='Points', by='Position', grid=False)

plt.title('Обобщенные показатели распределения очков по занятому месту')
plt.suptitle('')
plt.xlabel('Позиция в таблице')
plt.ylabel('Количество очков')
```

```python
print(max(data[data["Position"] == 19]["Points"]))
```

Максимальное количество баллов, которое набирали команды и при этом вылетели в низший дивизион - 39, а следовательно чтобы гарантированно не вылететь, то необходимо набрать 40 баллов
