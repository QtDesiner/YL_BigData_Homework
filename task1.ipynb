```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('sessions.csv')

# Разбиение по корзинам
bins = [0, 60, 300, 600, 900, 1200, 1800, 3600, 7200]
labels = ['<1м', '1-5м', '5-10м', '10-15м', '15-20m', '20-30м', '30-60м', '1-2ч']
data['duration_bins'] = pd.cut(data['duration'], bins=bins, labels=labels)

# Визуализация разбиения
plt.figure(figsize=(10, 6))
data['duration_bins'].value_counts().sort_index().plot(kind='bar')
plt.title('Разбиение количества сессий по длительности')
plt.xlabel('Длительность сессии')
plt.ylabel('Количество сессий')
plt.show()
```
