```python
import pandas as pd
import matplotlib.pyplot as plt
```

Забитые очки

```python
data = pd.read_csv('eplleaguetables.csv')

# Создание ящика с усами
plt.figure(figsize=(10, 6))
data.boxplot(column='GoalsScored', by='Position', grid=False)
plt.title('Обобщенные показатели распределения очков по забитым голам')
plt.suptitle('')
plt.xlabel('Позиция в таблице')
plt.ylabel('Количество очков')

plt.show()
```

Наблюдается линейная зависимость, неравномерна на всём протяжении, после 10 места эта зависимость становится более пологой\
Также после 10 места размах значений становится меньше\
Количество выбросов минимально

Пропущенные очки

```python
plt.figure(figsize=(10, 6))
data.boxplot(column='GoalsConceded', by='Position', grid=False)
plt.title('Обобщенные показатели распределения очков по забитым голам')
plt.suptitle('')
plt.xlabel('Позиция в таблице')
plt.ylabel('Количество очков')

plt.show()
```

Также наблюдается линейная зависимоть, более равномерая, чем у забитых голов\
Размах значений высокий на всём протяжении\
Также количество выбросов больше, чем у забитых голов
